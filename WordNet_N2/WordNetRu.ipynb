```
from wiki_ru_wordnet import WikiWordnet
wikiwordnet = WikiWordnet()
```

```
synsets = wikiwordnet.get_synsets('собака')
for i in synsets:
  print('------------------------------------------------------------------------------------------------------------------------')
  print('#', i)
  print({w.definition() for w in i.get_words()});
  print('+', {w.lemma() for w in i.get_words()});
  print('*', {w.lemma() for w in hypernym.get_words() for hypernym in wikiwordnet.get_hypernyms(i)})
  print('x', {w.lemma() for w in hyponym.get_words() for hyponym in wikiwordnet.get_hyponyms(i)})
```
